```python
import pandas as pd
zomato_df = pd.read_csv("../data/zomato.csv", encoding='latin-1')
country_df = pd.read_csv("../data/Country-Code.csv")

zomato_df.head()
```

```python
df = zomato_df.merge(country_df, how='left', on='Country Code')
df.head()
```

```python
columns_to_drop = ['Address', 'Locality', 'Locality Verbose',
                'Switch to order menu', 'Currency', 'Rating color', 'Rating text']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
df['Cuisines'] = df['Cuisines'].str.lower().str.strip()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
print(df['Average Cost for two'].describe())
```

```python
q1 = df['Average Cost for two'].quantile(0.25)
q3 = df['Average Cost for two'].quantile(0.75)

iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

df = df[
    (df['Average Cost for two'] >= lower_bound) &
    (df['Average Cost for two'] <= upper_bound)
]

print(df['Average Cost for two'].describe())
```

```python
df = df[df['Average Cost for two'] > 0]
```

###### | Low    | 0 – 250    |
###### | Medium | 251 – 600  |
###### | High   | 601 – 1350 |


```python
def cost_category(cost):
    if cost <= 250:
        return 'Low'
    elif cost <= 600:
        return 'Medium'
    else:
        return 'High'

df['Cost Category'] = df['Average Cost for two'].apply(cost_category)
```

```python
df.to_csv("../data/clean_df.csv", index=False)
```
