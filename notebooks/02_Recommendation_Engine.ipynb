```python
import pandas as pd 
df = pd.read_csv("../data/clean_df.csv")
df 
```

```python
def filter_restaurants(df, selected_cuisines, cost_category, selected_country):
    filtered_df = df[
        df['Primary Cuisine'].isin(selected_cuisines) &
        (df['Cost Category'] == cost_category) &
        (df['Country'] == selected_country)
    ]

    return filtered_df
```

```python
def rank_restaurants(df):
    if df['Votes'].max() == 0:
        df['norm_votes'] = 0
    else:
        df['norm_votes'] = df['Votes'] / df['Votes'].max()

    df['norm_rating'] = df['Aggregate rating'] / 5 

    df['Score'] = (df['norm_rating'] * 0.7) + (df['norm_votes'] * 0.3)

    return df.sort_values(by='Score', ascending=False)
```

```python
df = rank_restaurants(df)
df
```

```python
def add_explanation(df):
    df['Explanation'] = df.apply(
        lambda row: f"Matched on {row['Primary Cuisine']} cuisine, {row['Cost Category']} budget in {row['Country']} with rating {row['Aggregate rating']}",
        axis=1
    )
    return df
```

```python
df = add_explanation(df)
df
```

```python
df.to_csv("../data/sorted_df.csv", index=False)
```
